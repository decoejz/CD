A distribuição teórica que será utilizada para a variável X é a exponencial.

$X$~$Exp(α)$

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

Sorteando o número aleatório alto e criando os números aleatórios em cima desse valor.

```python
exponencial = expon.rvs(scale=1/8, size=1000000, loc=0)
```

```python
media = exponencial.mean()
variancia = exponencial.var()

print("Média populacional:",media)
print("Variância populacional:",variancia)
```

### n = 2

```python
medias_amostrais = []
for i in range (10000):
    n = 2

    amostra = np.random.choice(exponencial, size=n)

    media_amostral = amostra.mean()
    
    medias_amostrais.append(media_amostral)
    
    print('Média amostral {0}: {1}'.format(i+1, media_amostral))
```

```python
medias_amostrais_linha = pd.Series(medias_amostrais)
plot = medias_amostrais_linha.plot.hist(title='Histograma',normed=True)
```

```python
media_das_amostras = medias_amostrais_linha.mean()
variancia_das_amostras = medias_amostrais_linha.var()

print("Média das amostras:",media_das_amostras)
print("Variância populacional:",variancia_das_amostras)
```

## Interpretando os valores

O resultado obtido mostra um histograma com um comportamento muito próximo a uma função exponencial, o que diverge do Teorema do Limite Central, que prevê uma distribuição normal para a distribuição das médias.

### n = 50

```python
medias_amostrais = []
X1_lista = []
X9_lista = []
for i in range (10000):
    n = 50

    amostra = np.random.choice(exponencial, size=n)
    
    X1 = amostra[0]
    X1_lista.append(X1)
    
    X9 = amostra[8]
    X9_lista.append(X9)
    
    media_amostral = amostra.mean()
    
    medias_amostrais.append(media_amostral)
    
medias_amostrais_linha = pd.Series(medias_amostrais)
plot = medias_amostrais_linha.plot.hist(bins=n,title='Histograma',normed=True)
```

```python
media_das_amostras = medias_amostrais_linha.mean()
variancia_das_amostras = medias_amostrais_linha.var()

print("Média das amostras:",media_das_amostras)
print("Variância populacional:",variancia_das_amostras)
```

## Interpretando os valores

Para n = 4 a distribuição das médias se aproxima de uma distribuição normal com assimetria à direita. Já para n = 20 e n = 50 as médias seguem praticamente que uma distribuição normal simétrica.

___
# Conclusão

Para valores n pequenos o TLC não pode ser comprovado empiricamente, já que a distribuição das médias não pode ser considerada uma normal simétrica. Entretanto, para valores n altos a distribuição se assemelha bastante de uma normal simétrica e, dessa forma, é possível comprovar o TLC através de simulações. Portanto, quanto maior for o valor de n, melhor é a visualização experimental do Teorema.

___
## Comprovando o TLC

```python
X1_linha = pd.Series(X1_lista)
plot = X1_linha.plot.hist(title='Histograma - X1',normed=True)
```

```python
X9_linha = pd.Series(X9_lista)
plot = X9_linha.plot.hist(title='Histograma - X9',normed=True)
```

Com uma pequena diferença entre os valores resultantes, o comportamento do histograma de X1 é igual ao comportamento do histograma de X9, que é igual ao comportamento do histograma da função de probabilidade escolhida, no caso, exponencial.

```python
#Calculando a correlação entre X1 e X9
media_X1 = X1_linha.mean()
desvpad_X1 = np.sqrt(X1_linha.var())

media_X9 = X9_linha.mean()
desvpad_X9 = np.sqrt(X9_linha.var())
```

```python
covariancia = 0

for i in range (len(X1_lista)):
    covariancia+=((X1_lista[i]-media_X1)*(X9_lista[i]-media_X9))
    
covariancia/=len(X1_lista)
```

```python
correlacao = covariancia/(desvpad_X1*desvpad_X9)
print('A correlação entre X1 e X9 vale {0}'.format(correlacao))
```

A correlação tem um valor muito próximo de zero, que indica que são valores independetes.
